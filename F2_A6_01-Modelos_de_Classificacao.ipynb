# Separando as Bases

Primeiramente devemos separar as bases em Treino e Teste.  
Geralmente esta divisão se concentra em 70% para Treino e 30% pra Teste (Boas práticas).

A maior parte vai para Treino porque é a base que a máquina tem que aprender, então quanto mais dados temos disponíveis para treino melhor será o resultado,  
ou seja, mais preciso.  

Tem algoritmos que são de problemas de classificação, algoritmos para criação de *score* de perfil de clientes, que já não é mais classificação. Então
